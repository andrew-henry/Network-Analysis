## Visualization of street centrality

* In graph theory and network analysis, indicators of centrality identify the most important vertices within a graph

```python
Fort_edge_centrality = nx.closeness_centrality(nx.line_graph(Fort_Lauderdale))
```

```python
Fort_Lauderdale_edge_values = [Fort_edge_centrality[edge + (0,)] for edge in Fort_Lauderdale.edges()]

norm = colors.Normalize(vmin=min(Fort_Lauderdale_edge_values)*0.8, vmax=max(Fort_Lauderdale_edge_values))

cmap = cm.ScalarMappable(norm=norm, cmap=cm.inferno)

ec = [cmap.to_rgba(cl) for cl in Fort_Lauderdale_edge_values]

fig, ax = ox.plot_graph(Fort_Lauderdale, bgcolor='k', axis_off=True, node_size=0, edge_color=ec, edge_linewidth=1.5, edge_alpha=1)
```
